# Importing The Libraries

```
# Import the pandas library and the numpy library
#
import pandas as pd
import numpy as np
```

# Loading the Dataset

```
# Loading the data from the csv file into a dataframe to be used
#
df_auto = pd.read_csv("Autolib_dataset (2).csv", delimiter = ',')
df_auto
```

# Previewing the Dataset

```
# Previewing the first 10 rows of the data
#
df_auto.head(10)
```

# Accessing Information About the Dataset

```
# Checking the datatypes of the columns in the dataset
#
print(df_auto.info())
```

```
# Checking for any null values in the dataset
#
print(df_auto.isnull().any())

print(df_auto.isnull().sum())
```

# Data Cleaning

```
# Irrelevant Data Observation: Dropping unneeded rows and columns
# Dropping rows that are not in the city of Paris over the month of April 2018.
#
rows_needed = df_auto[(df_auto["City"] == "Paris") & (df_auto["month"] == 4) & (df_auto["year"] == 2018)]
rows_reset = rows_needed.reset_index()
rows_reset
```

```
# Dropping the Cars column since it's equivalent to the "Bluecar counter" column all through the dataset.
# Dropping the "Displayed comment", "ID" and "Geo point" columns since no questions to be answered require these columns, thus not needed for the analysis
# Dropping the "Scheduled at" column since it only represents the "future" status which has no resources (Bluecar,Utilib,Utilib 1.4 and parking slots)
# 
rows_reset.drop(["Cars", "Displayed comment", "ID", "Geo point", "Scheduled at"], axis = 1, inplace = True)
rows_reset
```

```
# Dropping rows where the "Charging Status" column is "future" since they don't have resources
# 
rows_reset.drop(rows_reset.index[rows_reset["Charging Status"] == "future"], inplace = True)
rows_reset
```

```
# Dropping rows where the "Kind" column is "CENTER" since they also don't have resources
#
rows_reset.drop(rows_reset.index[rows_reset["Kind"] == "CENTER"], inplace = True)
rows_reset
```

```
# Dropping rows where the "Station type" is "subs_center" since they also don't have resources
#
rows_reset.drop(rows_reset.index[rows_reset["Station type"] == "subs_center"], inplace = True)
rows_reset
```

```
# Dropping rows where the "Status" column is "scheduled" since it reprsents those that have "future" rental status thus have no resources
#
rows_reset.drop(rows_reset.index[rows_reset["Status"] == "scheduled"], inplace = True)
rows_reset
```

```
# Dropping rows where the "Subscription status" column is "future" since it reprsents those that have "future" rental status thus have no resources
#
rows_reset.drop(rows_reset.index[rows_reset["Subscription status"] == "future"], inplace = True)
rows_reset
```

```
# Dropping rows where the "Rental status" is not "operational" or "broken" since they don't have resources
#
df_valid = rows_reset[(rows_reset["Rental status"] == "operational") | (rows_reset["Rental status"] == "broken")]
df_valid
```

```
# Checking, counting and dealing with any missing values in the dataset
#
print(df_valid.isnull().any())

print(df_valid.isnull().sum())
```

```
# Dropping any duplicate columns from the dataset
#
df_valid.duplicated()

df_valid.drop_duplicates()
```

```
# Fixing the messy columns in the dataset
#
df_valid.columns = df_valid.columns.str.lower().str.replace(' ','_')
df_valid
```

# Exporting the Clean Dataset

```
# Export the clean dataset to a csv file
#
df_valid.to_csv('auto.csv')
```

# Data Analysis: Blue cars

# Question 1

```
# What is the most popular hour for returning Blue cars?
# This is the time where there are no more parking slots in the station since most people have returned the rented cars.
#
url = 'auto.csv'

clean_data = pd.read_csv(url)
clean_data

popular_hour = clean_data[clean_data ["slots"] == 0].groupby(["bluecar_counter"])["hour"].value_counts()
popular_hour
```

# Question 2

```
# What is the overall most popular station?
# This is the station that has the highest number of Blue cars
#
overall_popular_station = clean_data[clean_data ["bluecar_counter"] != 0]["public_name"].value_counts().head(1)
overall_popular_station 
```

# Question 3

```
# What is the most popular station at the most popular picking hour?
# This is the most popular station at 2100 hrs since the most popular returning hour is the most convinient time for renting a car
#
most_popular_station = clean_data[clean_data ["hour"] == 21].groupby(["bluecar_counter"])["public_name"].value_counts().head(1)
most_popular_station
```

# Question 4

```
# What the overall most popular postal code for picking up Blue cars?
#
overall_popular_station = clean_data[clean_data ["bluecar_counter"] != 0]["postal_code"].value_counts().head(1)
overall_popular_station 
```

# Question 5

```
# What is the most popular postal code at the most popular picking hour?
#  
most_popular_station = clean_data[clean_data ["hour"] == 21].groupby(["bluecar_counter"])["postal_code"].value_counts().head(1)
most_popular_station
```
